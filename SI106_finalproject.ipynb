```python
import requests
import json
from IPython.display import Image
from IPython.display import Audio


class Song():
    def __init__(self, song_title, song_artist, song_artwork, song_audio):
        self.song_title = song_title
        self.song_artist = song_artist
        self.song_artwork = song_artwork
        self.song_audio = song_audio
        
class Playlist():
    def __init__(self):
        self.playlist = []
        
    def addSong(self, song):
        self.playlist.append(song)
        return 
    
    def displayPlaylist(self):
        print(" ")
        print("=== PLAYLIST ===")
        try:
            myfile = open('playlist.txt', 'r') 
            lines = list(myfile.readlines())
            counter = 0
            for line in lines:
                if counter % 3 == 0:
                    display(Image(data=line, embed=True, format='jpg'))
                elif counter % 3 == 1:
                    print(line)
                elif counter % 3 == 2:
                    display(Audio(line))
                counter += 1
            return
        except:
            return "file doesn't exist"
    
    def savePlaylist(self):
        outFile = open('playlist.txt', 'w')
        for song in self.playlist:
            outFile.write('{}\n'.format(song.song_artwork))
            outFile.write('{} by {}\n'.format(song.song_title, song.song_artist))             
            outFile.write('{}\n'.format(song.song_audio))
        outFile.close()


myPlaylist = Playlist()

while True: 
    print('MAIN MENU')
    print("- Type 'exit' to leave")
    print("- Type 'playlist' to see your playlist so far")
    print("- Otherwise, enter an artist name")
    print(" ")

    user_inp = input("Enter an artist (or exit/playlist): ")
    
    if user_inp == "exit":
        break
    elif user_inp == "playlist":
        myPlaylist.displayPlaylist()
    else:
        r = requests.get('https://itunes.apple.com/search?', params = {
              'term': user_inp, 'media': 'music'
              })
        result = json.loads(r.text)

        for song in result['results']:
            display(Image(song['artworkUrl100']))
            print(song['trackName'] + ' by ' + song['artistName'])
            display(Audio(song['previewUrl']))
            user_inp2 = input("Add to playlist? (yes/no/back): ")
            if user_inp2 == "yes":
                thisSong = Song(song['trackName'], song['artistName'], song['artworkUrl100'], song['previewUrl'])
                myPlaylist.addSong(thisSong)
                myPlaylist.savePlaylist()
            elif user_inp2 == 'no':
                continue
            elif user_inp2 == 'back':
                break
```
